```python
#PCA on wine dataset.

import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine = load_wine();
df = pd.DataFrame(wine['data'], columns=wine['feature_names']); df

print(df.iloc[:,0:10].head())

print('*****Target******')
print(wine['target'])
```

```python
pca = PCA(n_components=2)

pca_model = pca.fit(df)
df_transform = pd.DataFrame(pca_model.transform(df), columns = ['pca1','pca2'])
df_transform
```

```python
# plot based on the dataframe

plt.scatter(df_transform.pca1, df_transform.pca2, c=wine.target, alpha = 0.8)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```
