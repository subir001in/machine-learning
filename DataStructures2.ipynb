#DICTIONARIES

## How Dict are created and used in Python.

## Each value in a dict is an immutable combination of key-value pair.
## Key is immutable but the value can be any mutable/immutable Python object.

```python
D = {'c' : 'India' , 'i' : [1,2,3]}; D
```

```python
## Pass list of tuples inside the dict function. Tuples have yje key-value pairs.
D1 = dict([('c' , 'India'), ('i' , [1,2,3]) , ('f' , [1,2,2.3,4.5])]) ; D1
```

## Accessing Dict values using the keys.

```python
D['i']
```

```python
D1['f']
```

## Using assignment operator to add elements to the Dict.

```python
D['f'] = [1,2,2.3,4.5];  D
```

```python
D['dummy'] = 'xyz'; D[9] = ' a value'; D
```

# DICT AND LIST COMPREHENSIONS



*   Basic Form: [expr for val in collection if condition]


```python
## List Comprehension
squared = [a*a for a in (1,2,3) if a >= 2 ] ; squared
```

```python
l = ['India' , 'is' , 'Bharat'] ; l
```

```python
L = [w.upper() for w in l ] ; L
```
