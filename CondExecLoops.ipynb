#LOOPS -- Used to iterate over the collection.

## FOR LOOPS

```python
for x in range(1,4):
  print(x)
```

```python
#checking the odd nos...
for x in range(1,10):
  if x%2 == 0:
    continue
  print(x,' is an odd number...')
```

```python
#checking prime nos

y=10
for x in range(2,y):
  if(y % x == 0):
    print(y,' is not a prime no.')
    break;
```

## WHILE LOOP

```python
x=32; c=0;
while x % 2 == 0:
  x = x/2
  c += 1
print(c,' times divisible by 2.')
```
