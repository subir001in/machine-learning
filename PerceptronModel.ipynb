```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data[:,(2,3)]
y = (iris.target==0).astype(np.int32)
```

```python
X.shape
```

```python
DF = pd.DataFrame(X, columns=('petal length', 'petal width'))
DF.info()
DF
```

```python
# Visualise the data

plt.plot(X[y==0,0], X[y==0,1], 'bs', label = 'Not Iris Setosa')
plt.plot(X[y==1,0], X[y==1,1], 'go', label = 'Iris Setosa')

plt.xlabel('Petal Length', fontsize=14)
plt.ylabel('Petal Width', fontsize=14)

plt.legend(loc='lower right', fontsize=14)
plt.show()
```

## Since the above datapoints can be clearly classified, the perceptron model can clearly do teh classification.

```python
# Apply perceptron model

from sklearn.linear_model import Perceptron

per_clf = Perceptron(max_iter=1000, tol=1e-3, random_state = 42 )
```

```python
per_clf.fit(X,y)
```

```python
y_pred = per_clf.predict([[2,0.5]])
```
