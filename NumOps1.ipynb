# FANCY INDEXING
##ARRAY Filtering and Dicing using Fancy Indexing

*   Basic Form: [[1st dim]] [:,[2nd]] [:,:,[3rd]]
*   Incase some dimensions are not needed then they can be Omitted, all the elements for the Omitted dimension can be selected.



```python
A_8_4
```

##To Select Rows from a 2D Array, specify the desired Indexes for 1st dimension

```python
#Slice the Array to Select the row 1,3,5 from the Array.

A_8_4[[1,3,5]]
```

##To Select columns from a 2D Array, specify the desired Indexes for 2nd dimension

```python
A_8_4
```

```python
#Slice the Array to select the 1st and 3rd COLUMNS
A_8_4[:,[1,3]]
```

## Similarly Select both ROWS and COLUMNS in 2D Array

```python
#Select both rows and cols in 2D Array
A_8_4[[1,3,5]][:,[1,3]]
```
