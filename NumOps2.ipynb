##Inner Matrix Product van be calculated by any of the below methods:-

*   np.dot() method
*   @



```python
A_6_3 = np.random.randn(6,3)
```

```python
A_3_6 = np.random.randn(3,6)
```

```python
#Multiply the matrix
np.dot(A_3_6,A_6_3)
```

##Compute INVERSE OF a MATRIX using 

*   np.linalg.inv() method



```python
A_5_5 = np.random.randn(5,5) ; A_5_5
```

```python
#Inverse of the matrix
np.linalg.inv(A_5_5)
```

```python
#Compute QR Decoposition of the matrix.
q, r = np.linalg.qr(A_5_5);
```

##Vectorized version of Terniary Operation.

*   Vectorized version is similar to the List Comprehension.
*   np.where(condn arg , argument for TRUE , argument for FALSE)
*   Return elements chosen from x or y depending on condition.



```python
np.where(np.array([True, False, False, True, False]), np.array([1.1, 1.2, 1.3, 1.4, 1.5]), np.array([2.1, 2.2, 2.3, 2.4, 2.5]))
```
