**SKITLEARN LIBRARY**

In skitlearn, all objects share a consistent and simple interface.

ESTIMATORS
------------------
Any object that can estimate or learn some of the model parameters , given a dataset.

 > FIT() METHOD --> Performs the Estimation process.
    -- Access learned parameters via public instance variables with underscore suffix.
 
 > HYPERPARAMETERS
    -- Access Hyperparameters via public instance variables.

TRANSFORMERS
------------------
A type of estimator that can transform a dataset.

  > TRANSFORM() METHOD --> transform() method transforms the dataset using learned parameter.

  > FIT_TRANSFORM() METHOD --> fit_transform() method is an optimized method to do both fit and transform.


PREDICTORS
------------------
A type of estimator, that can make predictions given a dataset.

  > PREDICT() METHOD --> predict() method takes the dataset and returns the prediction.
  
  > SCORE() METHOD --> score() method to measure the quality of predictions.
