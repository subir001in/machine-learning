```python
# K-Means Clustering Technique -- Implementation 2
# This example shows a demonstration where the KMeans do not work in a very efficient manner.

from sklearn.datasets import make_moons

X, y = make_moons(200, noise = 0.05, random_state = 0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], s=50, cmap = 'virdis')

```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, max_iter=10000)

kmeans.fit(X)
labels = kmeans.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=labels)
```
