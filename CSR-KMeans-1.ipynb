```python
# Illustration of K-Means on synthetic data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.6)

plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, max_iter=10000)
kmeans.fit(X)
labels = kmeans.predict(X)
```

```python
plt.figure(figsize = (7,5))
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

```python
# Deciding optimal number of clusters.

from sklearn.metrics import davies_bouldin_score

sse, db = {}, {}

for k in range(2,10):
  # seed of 10 for reproducibility.
  kmeans = KMeans(n_clusters = k, max_iter = 1000, random_state = 10).fit(X)

  sse[k] = kmeans.inertia_
  db[k] = davies_bouldin_score(X, labels)
```

```python
# Elbow method

plt.figure(figsize=(7,5))
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of Cluster")
plt.ylabel("SSE")
plt.show()
```

```python
# Davis-Boldin Index

plt.figure(figsize=(7,5))
plt.plot(list(db.keys()), list(db.values()))
plt.xlabel("Number of Cluster.")
plt.ylabel("Davies-Bouldin values")
plt.show()
```

```python
# Accuracy Metrics
# Calculate Silhoutte Score
from sklearn.metrics import silhouette_score
score = silhouette_score(X, labels, metric = 'euclidean')

# Print the score 
print("Silhoutter Score: %.3f" % score)
```

```python
# Accuracy metrics:
# Calculate homogeniety score:
from sklearn.metrics import homogeneity_score
A=homogeneity_score(y_true, labels)
print(A)
```
