#Line Graphs and Scatter Plots using Matplotlib, Seaborn and pandas plotting functionality

##Plot using Matplotlib

```python
import numpy as np
import pandas as pd
```

```python
TS = pd.Series(np.random.randn(1000).cumsum(), index = pd.date_range("1/1/2021", periods = 1000)); TS
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(TS)
plt.xticks(rotation = 90)
plt.xlabel('Day')
plt.ylabel('finance index')
plt.title('Changes in finance index using daily data')
plt.grid(visible=True)
```

##Plot using Pandas Plotting

```python
# Note in Pandas plotting you don't have to adjust the xticks as we did earlier.

TS.plot(xlabel='Day' , ylabel='Finance Index', title = 'Changes in finance data using daily data', grid = True)
plt.show()
```

##Plot using Seaborn

```python
import seaborn as sns
```

```python
sns.lineplot(x=TS.index, y=TS.values)
plt.xticks(rotation = 90)
plt.xlabel('Day')
plt.ylabel('finance index')
plt.title('Changes in finance index using daily data')
plt.grid(visible=True)
```

##Draw Scatter plots using some Polynomial data.

```python
x = 6 * np.random.randn(100) - 3
```

```python
y = 0.5 * x ** 2 + x + 2 + np.random.randn(100)
```

```python
plt.plot(x,y,'b.')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

##Draw Scatter plot using Pandas Plotting.

```python
DF = pd.DataFrame({'x':x , 'y':y}); DF
```

```python
DF.plot.scatter('x','y')
#DF.plot('x','y',kind='scatter') # This stmt also plots the same graph as above stmt.
plt.show()
```

```python
sns.scatterplot(data = DF,x='x',y='y')
plt.show()
```

#BAR CHART AND HISTOGRAMS

```python
import numpy as np
```

```python
x = [1,2]
```

```python
y = np.random.randn(2)
```

```python
tick_label = ['one','two']
```

```python
plt.bar(x,y,tick_label = tick_label)
plt.xlabel('group')
plt.ylabel('height')
plt.show()
```

## Do the same plotting using the Pandas plotting.

```python
x=['one','two']
```

```python
DF = pd.DataFrame({'x': x , 'y' : y})
```

```python
DF.plot.bar('x','y',xlabel='group',ylabel='height',legend=False,rot=0)
plt.show()
```

##Use seaborn plotting for plotting Bar Chart

```python
sns.barplot(x='x',y='y', data = DF)
plt.xlabel('group')
plt.ylabel('height')
plt.show()
```

##HISTOGRAM

```python
x = np.random.normal (0,1,size=100);
```

```python
plt.hist(x,rwidth = 0.95)
#plt.xlabel('variable')
#plt.ylabel('frequency')
plt.show()
```

##Use Pandas Plotting for Histogram

```python
DF = pd.DataFrame({'x':x});
```

```python
DF['x'].plot.hist(rwidth = 0.95,xlabel='variable',ylabel='frequency')
plt.show()
```

##Draw Histogram using Seaborn

```python
sns.histplot(data=DF, x=DF['x'],)
plt.xlabel('variable')
plt.ylabel('frequency')
```

#BOXPLOTS, and MULTIDIMENSIONAL VISUALIZATION 

```python
x = np.random.normal(0,1,size=100)
```

```python
plt.boxplot(x, showmeans = True)
plt.xlabel('distribution')
plt.ylabel('variable')
plt.show()
```

##Plot using pandas...

```python
DF = pd.DataFrame({'x':x})
```

```python
DF['x'].plot.box(showmeans= True, xlabel= 'distribution', ylabel='variable')
```

```python
sns.boxplot(data=DF,x=x)
plt.xlabel('distribution')
plt.ylabel('variable')
plt.show()
```

## MULTI-DIMENSIONAL VISUALIZATION -- To add multi-dimensional visualisation, add 3rd dimensional visual color.

```python
x = 6 * np.random.randn(100) - 3
```

```python
y = x + 2 + np.random.randn(100)
```

```python
x1 = np.random.randint(0,2,size=100);
```

```python
plt.scatter(x,y,c=x1,cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='x1')
plt.show()
```

## Scatter Plot using Pandas

```python
DF = pd.DataFrame({'x':x, 'y':y, 'c':x1},index=range(100));DF
```

```python
DF.plot.scatter('x','y',c='c' , colormap='viridis', legend = True, xlabel ='x',ylabel ='y',sharex=False)
plt.show()
```

```python
sns.scatterplot(x='x',y='y',hue='c' ,data=DF)
plt.show()
```

##MULTI-PANEL SUBPLOTS

```python
TS1 = pd.Series(np.random.randn(1000).cumsum() , index = pd.date_range("1/1/2020", periods=1000));
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(TS1)
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Index')
axes[0,0].grid()
axes[0,0].tick_params(axis='x', labelrotation=90)

plt.tight_layout()
```

##Draw Multipanel Subplots using Pandas

```python
fig, axes = plt.subplots(2,2)

TS1.plot(ax=axes[0,0],xlabel='Month', ylabel='Index', grid=True)
plt.show()
```
