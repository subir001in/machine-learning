#REGULAR EXPRESSIONS

###Basic structure of Regex expression has 2 components :

*   *Metacharacters*( *Metacharacters* provide pattern creation functionality.)

1.   [] - used to specify a set of characters without any special meaning.
2.   \* - used to signal zero or more occurances of a set of characters.
3.   \+ - used to signal one or more occurances of a set of characters.
4.   {} - used to specify the number of occurances that we want for a set of characters.
5.   () - used to capture and group characters.



*   Special sequences( *Special sequences* provide specific pattern matching functionality. They are written with \\ character.)

1.   \\s - used to find a match where a string contains a whitespace character.


*   Common Regex methods: 

1.   split() -- Splits the input string based on the regex pattern string.
2.   findall() -- Finds all the regex character/s inside the input string based on the regex pattern string.













```python
import re
```

```python
text1 = '''India is Bharat
Bharat is 
India'''
```

```python
re.split('\s+',text1)
```

## Steps to Improve the re-usability of the Regex or regular expression.

*   use *compile()* method to First compile the Regex pattern to create the regex object.
*   Apply the required method to the regex object. 



```python
#Step 1: Create the regex object.
regex = re.compile('\s+')

#Step 2: Apply the required method to the regex object.
regex.split(text1)
```

```python
regex.findall(text1)
```

##Identify Username from email address using Regex expression

```python
email = 'subir-samantaray@gmail.com'
pattern = r'([A-Za-z0-9._%-]+)@'
regex = re.compile(pattern)
```

```python
regex.findall(email)
```
