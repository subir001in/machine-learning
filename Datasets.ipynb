##Perform some Data Analytics to Compute correlations between 2 columns. 

```python
#Computing correlation between the Mediam House value and the Median Income
df1['median_house_value'].corr(df1['median_income'])
```

```python
#Computing pairwise correlation with one particular column of interest with all other columns in a Dataframe.
```

```python
df1.corrwith(df1.median_house_value)
```

```python
df1.corrwith(df1.median_house_value).sort_values(ascending=False)
```

##Compute Co-variance between 2 columns, use the *cov( )* method.

```python
df1['median_house_value'].cov(df1['median_income'])
```
