```python
Score = pd.read_csv('student_scores.csv')
Score.shape
Score
```

```python
# Plotting the data.

Score.plot(x='Hours', y='Scores', style = 'o')
plt.xlabel='Hours'
plt.ylabel='Scores'
plt.show()
```

```python
# Preparaing the data

X = Score.iloc[:,:-1].values
y = Score.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X, y)
```

```python
# printing parameters

print(reg.intercept_)
print(reg.coef_)
```

```python
# making predictions

y_pred = reg.predict(X)

df = pd.DataFrame({'Actual':y , 'Predicted':y_pred})
df
```

```python
# The co-efficients

from sklearn.metrics import mean_squared_error, r2_score
# The mean squared error

print('Mean squared error : %.2f' % mean_squared_error(y, y_pred))

# The co-effiecient of determination: 1 is a perfect prediction.

print('R2 score : %.2f' % r2_score(y, y_pred))

# plot outputs

plt.scatter(X, y, color = 'black')
plt.scatter(X, y_pred, color = 'blue', linewidth = 2)

plt.xticks(())
plt.yticks(())
plt.show()
```
