# Apply Principal Component Analysis(PCA) on iris dataset. Reduce 4 features to 2 features.

```python
# Importing Utilities:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
# Importing datasets
iris = datasets.load_iris()

X = iris.data; 
y = iris.target;

print('******** Input Features *********')
print(X[0:5])

print('******** Labels *********')
print(y)
```

```python
# Reducing Dimensions..

pca = PCA(n_components = 2)
pca.fit(iris.data)
pca.transform(iris.data)

plt.scatter(X[:, 0], X[:, 1], c = iris.target)
```
