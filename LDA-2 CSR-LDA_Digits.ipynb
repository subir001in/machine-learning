```python
#import digits data

from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
```

```python
#Loading data

digits = load_digits(); 
digits.data.shape
```

```python
X = digits.data
y = digits.target

lda = LDA(n_components = 2)
New = lda.fit_transform(X, y)

New.shape
```

```python
# Classification using LDA

xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.33)

lda_model = lda.fit(xtrain, ytrain)

yn = lda.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(ytest, yn)
```

```python
# Get the classification report

from sklearn.metrics import classification_report

target_names = ['d0','d1','d2','d3','d4','d5','d6','d7','d8','d9']
print(classification_report(ytest, yn, target_names=target_names))
```
